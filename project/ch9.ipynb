## 객체지향 인터페이스

1. plt.figure
2. plt.subplot
3. plt.axes()

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([1,2,3,4], [10, 20, 25, 30], color = 'blue', linestyle = '-', linewidth = 2, marker = 'o')
# ax.plot([1,2,3,4], [20, 25, 30], color = 'red')

ax.set_title('sample plot')
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.plot([1,2,3,4], [10, 20, 25, 30], color = 'blue', linestyle = '-', linewidth = 2)
ax1.set_title('sample plot')


ax2 = fig.add_subplot(2,2,2)
ax2.plot([1,2,3,4], [10, 20, 25, 30], color = 'red', linestyle = '-', linewidth = 2)
ax2.set_title('sample2 plot')

ax3 = fig.add_subplot(2,2,3)
ax3.plot([1,2,3,4], [10, 20, 25, 30], color = 'green', linestyle = '--', linewidth = 2)
ax3.set_title('sample3 plot')


ax4 = fig.add_subplot(2,2,4)
ax4.plot([1,2,3,4], [10, 20, 25, 30], color = 'purple', linestyle = '--', marker = "^", linewidth = 2)
ax4.set_title('sample4 plot')

plt.tight_layout()
plt.show
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot([1,2,3,4], [10, 20, 25, 30], color = 'red', linestyle = '-', linewidth = 2)

plt.scatter([2,3,4], [20,25, 30], color = 'red')

ax.set_title("Plot Test")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.scatter(x, y, color = 'blue')
ax1.set_title("Main Plot")
ax1.set_xlabel("Random X")
ax1.set_ylabel("Ramnom Y")

ax2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
ax2.hist(y, bins = 10, color = 'green')

ax2.hist(y, bins = 20, color = 'green')
ax2.set_title("sub_Insert Plot")
ax2.set_xlabel("Random X")
ax2.set_ylabel("Frenquecy Y")
```

```python
fig, axarr = plt.subplots(2,2)


axarr[0,0].plot([1,2,3,4], [10, 20, 25, 30], color = 'blue', linestyle = '-', linewidth = 2)
axarr[0,0].set_title('sample plot')


axarr[0,1].plot([1,2,3,4], [10, 20, 25, 30], color = 'red', linestyle = '-', linewidth = 2)
axarr[0,1].set_title('sample2 plot')

axarr[1, 0].plot([1,2,3,4], [10, 20, 25, 30], color = 'green', linestyle = '--', linewidth = 2)
axarr[1, 0].set_title('sample3 plot')


axarr[1,1].plot([1,2,3,4], [10, 20, 25, 30], color = 'purple', linestyle = '--', marker = "^", linewidth = 2)
axarr[1,1].set_title('sample4 plot')

plt.tight_layout()
plt.show
```
